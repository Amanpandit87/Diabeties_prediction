```python
df = df[(df["BMI"] > 10 ) & (df["BMI"] < 50)]
```

```python
df.shape
```

```python
x = df.drop(columns = ["Outcome"],axis = 1)
y = df["Outcome"]
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train = std.fit_transform(x_train)
x_test = std.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
y_pred = lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("LogisticRegression = ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion='entropy',max_depth = 5)
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rd = RandomForestClassifier(criterion='entropy',
    max_depth=5)
```

```python
rd.fit(x_train,y_train)
```

```python
y_pred = rd.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB
```

```python
nv = GaussianNB()
```

```python
nv.fit(x_train,y_train)
```

```python
y_pred = nv.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
brnb = BernoulliNB()
```

```python
brnb.fit(x_train,y_train)
```

```python
y_pred = brnb.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag = BaggingClassifier()
```

```python
bag.fit(x_train,y_train)
```

```python
y_pred = bag.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
estimators = [("rd",rd),("lg",lg),("dt",dt),("svc",svc),("nv",nv),("brnb",brnb),("bag",bag)]
```

```python
from sklearn.ensemble import StackingClassifier
```

```python
st = StackingClassifier(estimators = estimators,final_estimator = rd)
```

```python
st.fit(x_train,y_train)
```

```python
y_pred = st.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```
